```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
train_df = pd.read_csv('/content/train.csv')
test_df = pd.read_csv('/content/test.csv')
```

```python
print(train_df.shape)
print(test_df.shape)

train_df.head()
```

```python
X_train = train_df[['GrLivArea', 'BedroomAbvGr', 'FullBath']]
y_train = train_df['SalePrice']
```

```python
model = LinearRegression()

model.fit(X_train, y_train)
```

```python
print("Intercept:", model.intercept_)

for feature, coefficient in zip(X_train.columns, model.coef_):
    print(feature, ":", coefficient)
```

```python
X_test = test_df[['GrLivArea', 'BedroomAbvGr', 'FullBath']]
```

```python
predictions = model.predict(X_test)

print(predictions[:10])
```

```python
submission = pd.DataFrame({
    'Id': test_df['Id'],
    'PredictedSalePrice': predictions
})

submission.head()
```

```python
submission.to_csv('predictions.csv', index=False)

print("Prediction file created successfully!")
```
